```python
import turtle
import math
from random import choice

#Fondo juego
wn=turtle.Screen()
wn.bgcolor("black")


#Limite del juego 
border_pen=turtle.Turtle()
border_pen.speed(0)
border_pen.color("white")
border_pen.penup() #No permite que el turtle dibuje sobre el plano
border_pen.setposition(-300,-300)
border_pen.pendown()
border_pen.pensize(3)
for side in range(4):
    border_pen.fd(600)
    border_pen.lt(90)
border_pen.hideturtle()

#Create a player
player=turtle.Turtle()
player.hideturtle()
player.color("white")
player.shape("square")
player.shapesize(0.2,4)
player.penup()
player.speed(0)
player.setposition(0,-275)
player.showturtle()

playermove=10

#Create a ball
ball=[0,0,0]
b=0
ball[b]=turtle.Turtle()
ball[b].hideturtle()
ball[b].color("white")
ball[b].shape("circle")
ball[b].shapesize(0.2,0.2)
ball[b].penup()
ball[b].speed(0)
ball[b].setposition(0,0)
ball[b].showturtle()
#start=False
    
#Funcion crear otra bola
def Create_a_ball():
    ball[b]=turtle.Turtle()
    ball[b].hideturtle()
    ball[b].color("white")
    ball[b].shape("circle")
    ball[b].shapesize(0.2,0.2)
    ball[b].penup()
    ball[b].speed(0)
    ball[b].setposition(0,0)
    ball[b].showturtle()
    #start=False

#Escribir un rand entre -1 y 1 multiplicando a cada componente de ballspeed 
#con un if si es cero, faltaria un for
ballspeedx=[2,2,2]
ballspeedy=[2,2,2]


for h in range(3):
    #ballspeedx[h]*=random.randrange(-1,1,1)
    #ballspeedy[h]*=random.randrange(-1,1,2)
    ballspeedx[h]*=choice((1,-1))
    ballspeedy[h]*=choice((1,-1))

    
    
    

#Create a block to deto
i=0
j=0
f=int(input("Introduzca numero de filas: "))
block=[0,0,0,0,0,0,0,0]*f
for j in range(f):
    for i in range(8):
        block[i+8*j]=turtle.Turtle()
        block[i+8*j].hideturtle()
        block[i+8*j].color("white")
        block[i+8*j].shape("square")
        block[i+8*j].shapesize(0.4,2.7)
        block[i+8*j].penup()
        block[i+8*j].speed(0)
        block[i+8*j].setposition(-262+(i*75),275-(j*25))
        block[i+8*j].showturtle()
        



#Move the player left and right
def move_left():
    x=player.xcor()
    x -= playermove
    if x < -280:
        x = -280
    player.setx(x)
    
def move_right():
    x=player.xcor()
    x += playermove
    if x > 280:
        x = 280
    player.setx(x)
    
#Distance function of colision
def Colision(var1,var2):
    #distance = math.sqrt(math.pow(var1.xcor()-var2.xcor(),2)+math.pow(var1.ycor()-var2.ycor(),2))
    if var2.xcor()-30 < var1.xcor() and var1.xcor() < var2.xcor()+30 and var1.ycor()>var2.ycor()-10:
    #if distance <10:
        return True
    else:
        return False
    
#Funcion pierdes 
def derrota():
    turtle.hideturtle()
    turtle.pu()
    turtle.color("red")
    turtle.goto(0,0)
    turtle.write("Has perdido",False,"Center",("Arial", 100, "normal"))
    
#Funcion ganas 
def victoria():
    turtle.hideturtle()
    turtle.pu()
    turtle.color("red")
    turtle.goto(0,0)
    turtle.write("Has ganado",False,"Center",("Arial", 100, "normal"))


    
#Create a keyboard bindings
turtle.listen()
turtle.onkey(move_left,"Left")
turtle.onkey(move_right,"Right")

#contador_de_blocks=0
#numero_blocks=len(block)

m=0
bolas_perdidas=0
#start=True

while True:
    
    #if contador_de_blocks==numero_blocks:
        #victoria()
      
    
    #Moverse la bola
    ballx=ball[b].xcor()
    bally=ball[b].ycor()
    ballx+=ballspeedx[b]
    bally-=ballspeedy[b]/2
    #ball.setx(ballx)
    #ball.sety(bally)
    ball[b].setposition(ballx,bally)
    
    #Limites de bola eje x
    if ball[b].xcor() > 290:
        ballspeedx[b]*=-1

    if ball[b].xcor() < -290:
        ballspeedx[b]*=-1

    #Limites de bola eje x
    if ball[b].ycor() > 290:
        ballspeedy[b]*=-1

    if ball[b].ycor() < -290:
        ballspeedy[b]=0
        ballspeedx[b]=0
        bolas_perdidas+=1
        #ball.hideturtle()
        #ball[b].setposition(300,300)
        #Start=True
        if bolas_perdidas == 3:
            derrota()
            break
        else:
            b+=1
            Create_a_ball()
        
        
        #derrota()  #Descomentar ambos derrota y break
        #break
        
    #Rebote de la bola con el player
    if player.xcor()-50 < ball[b].xcor() and ball[b].xcor() < player.xcor()+50 and ball[b].ycor()<player.ycor()+10:
        ballspeedy[b]*=-1
        
    #Colision de la bola con los blocks
    #aux=True
    #if True:
    salir=False
    for j in range(f):
        for i in range(8):
            if Colision(ball[b],block[i+8*j]): #and aux:
                #Se esconde el bloque
                block[i+8*j].hideturtle()
                block[i+8*j].setposition(350,350)
                #Rebota la bola
                #bally=ball.ycor()-15
                #ball.sety(bally)
                ballspeedy[b]*=-1
                #contador_de_blocks+=1
                #aux=False
                salir=True
                break
        if salir:
            break
        
            
            


    


turtle.done()
```
